# EU-SILC Country Report
Generated with papermill.

```python
COUNTRY = 'LU'
YEARS = list(range(2008, 2014))
TITLE = f'EU-SILC: {COUNTRY}'
```

```python
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
def find_project_root(s):
    for c in [s, *s.parents]:
        if (c / 'pyproject.toml').exists(): return c
    raise FileNotFoundError()
PROJECT_ROOT = find_project_root(Path().resolve())
os.chdir(PROJECT_ROOT)
DATA_DIR = PROJECT_ROOT / "data"
from silc_toolkit import load_incomes, arop_rate, gini_coefficient, s80s20_ratio
print(f"Report: {TITLE}")
```

```python
records = []
for year in YEARS:
    try:
        inc = load_incomes(COUNTRY, year, DATA_DIR)
        if inc:
            inc_s = sorted(inc); n = len(inc_s)
            records.append({"Year": year, "N": n,
                "Median": round(inc_s[n//2],0),
                "AROP%": round(arop_rate(inc)*100,1),
                "Gini": round(gini_coefficient(inc),4),
                "S80S20": round(s80s20_ratio(inc),2)})
    except FileNotFoundError: pass
df = pd.DataFrame(records)
print(df.to_string(index=False))
```

```python
from IPython.display import Image, display
if len(df) > 0:
    fig, axes = plt.subplots(1, 3, figsize=(13, 4))
    for ax, col, label, clr in zip(axes,
        ["AROP%", "Gini", "S80S20"],
        ["AROP (%)", "Gini", "S80/S20"],
        ["steelblue", "coral", "seagreen"]):
        ax.plot(df["Year"], df[col], marker="o", color=clr)
        ax.set_title(f"{COUNTRY} — {label}")
        ax.grid(True, alpha=0.3)
    plt.suptitle(TITLE); plt.tight_layout()
    out = PROJECT_ROOT / "reports" / f"{COUNTRY}_trend.png"
    plt.savefig(out, dpi=100, bbox_inches="tight"); plt.close()
    display(Image(filename=str(out)))
    print(f"Chart saved: {out.name}")
```
